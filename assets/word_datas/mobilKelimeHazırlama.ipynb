```python
import PyPDF2

with open('The_Oxford_3000_merged.pdf', 'rb') as file:
    reader = PyPDF2.PdfReader(file)
    text = ''
    
    for page_num in range(len(reader.pages)):
        page = reader.pages[page_num]
        text += page.extract_text()

print(text)
```

```python
import re

pattern = r'(\w+)\s\w+.\s(A1|A2|B1|B2|C1|C2)' 
matches = re.findall(pattern, text)

word_list = [{'word': match[0], 'level': match[1]} for match in matches]
word_list_sorted = sorted(word_list, key=lambda x: x['level'])
json_output = json.dumps(word_list_sorted, indent=4)

print(json_output)
print(len(json.loads(json_output)))
filename = 'english_words.json'

with open(filename, 'w', encoding='utf-8') as json_file:
    json_file.write(json_output)

print(f"Data saved to {filename}")
```

```python
import json
with open(path + "english_words_level_id.json", "r", encoding="utf-8") as file:
    data = json.load(file)

levels = {
    "A1": [],
    "A2": [],
    "B1": [],
    "B2": [],
    "C1": []
}
for word_data in data:
    level = word_data["level"]
    if level in levels:
        levels[level].append(word_data)
        
for level, words in levels.items():
    for index, word_data in enumerate(words, start=1):
        word_data["id"] = index
        
    filename = path + f"english_words_{level.lower()}.json"
    with open(filename, "w", encoding="utf-8") as outfile:
        json.dump(words, outfile, indent=4, ensure_ascii=False)
```
